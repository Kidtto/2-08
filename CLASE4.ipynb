```python
class OperacionesCadena:



    def  __init__(self,cadena):

        self.__cadena = cadena;





    def  set_cadena(self, cadena):

        self.__cadena = cadena;





    def  get_cadena(self ):

        return self.__cadena;





    def  obtener_longitud_cadena(self ):

        return len(self.__cadena);





    def  convertir_mayusculas(self ):

        return self.__cadena.upper();



    def  convertir_minusculas( self):

        return self.__cadena.lower();



    def  imprimir_caracteres_cadena( self ):

        for c in self.__cadena:

            print (c)





def run():



    c = input('Digite una cadena ')

    cadena = OperacionesCadena(c)

    print('Longitud de la cadena',cadena.obtener_longitud_cadena())

    print('Convertir a mayúsculas la cadena',cadena.convertir_mayusculas())

    print('Convertir a minusculas la cadena',cadena.convertir_minusculas())

    print('Imprimir caracteres de la cadena',cadena.imprimir_caracteres_cadena())







if __name__ == '__main__':

    run()
```

```python
#METODOS PARA VALIDACIÓN DE DATOS TIPO STRING



#isdigit(): Devuelve True si la cadena es todo números (False en caso contrario)

c = "-100"

print("-100",c.isdigit())



#isalpha(): Devuelve True si la cadena es todo carácteres alfabéticos

a = "ABC10034po"

b = "Holamundo"

print("ABC10034po ",a.isalpha())

print("Holamundo ",b.isalpha())



#isalnum(): Devuelve True si la cadena contiene números o carácteres alfabéticos

c = "ABC10034po"

print("ABC10034po ",c.isalnum())



#istitle(): Devuelve True si la primera letra de cada palabra es mayúscula:

c = "Hola Mundo"

print("Hola Mundo ",c.istitle())



#isspace(): Devuelve True si la cadena es todo espacios

c = "   "

print( "   ",c.isspace())



#startswith(): Devuelve True si la cadena empieza con una palabra en especifico

c= "Hola mundo"

print("Hola mundo ",c.startswith("Hola"))



#endswith(): Devuelve True si la cadena termina con una palabra en especifico

c= "Hola mundo"

print("Hola mundo ",c.endswith("mundo"))
```
